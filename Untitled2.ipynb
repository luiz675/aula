```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/guilhermeonrails/data-jobs/refs/heads/main/salaries.csv")

```

```python
df.head(10)
```

```python
df_limpo = df.dropna()
```

```python
df_limpo.isnull().sum()
```

```python
df_limpo.head()
```

```python
df_limpo['experience_level'].value_counts().plot(kind='bar', title='Distruibuição de senioridade')
```

```python
import seaborn as sns
```

```python
sns.barplot(data=df_limpo, x='experience_level', y='salary_in_usd')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df_limpo, x='experience_level', y='salary_in_usd')
plt.title('Salario médio por senioridade')
plt.xlabel('senioridade')
plt.ylabel('salário média anual (salary_in_usd)')
plt.show
```

```python
df_limpo.groupby('experience_level')['salary_in_usd'].mean().sort_values(ascending=False)
```

```python
ordem = df_limpo.groupby('experience_level')['salary_in_usd'].mean().sort_values(ascending=False).index
```

```python
ordem
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df_limpo, x='experience_level', y='salary_in_usd', order=ordem)
plt.title('Salario médio por senioridade')
plt.xlabel('senioridade')
plt.ylabel('salário média anual (salary_in_usd)')
plt.show
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df_limpo['salary_in_usd'], bins= 50, kde=True)
plt.title('Distribuição dos salário anuais')
plt.xlabel('Salário em usd')
plt.ylabel('Frequência')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df_limpo['salary_in_usd'])
plt.title('Boxplot dos salários')
plt.xlabel('Salário em usd')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='experience_level', y='salary_in_usd', data= df_limpo, order=ordem)
plt.title('Boxplot da Distribuição de Salário por Senioridade')
plt.xlabel('Senioridade')
plt.ylabel('Salário em USD')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='experience_level', y='salary_in_usd', data= df_limpo, order=ordem, palette='Set2', hue='experience_level')
plt.title('Boxplot da Distribuição de Salário por Senioridade')
plt.xlabel('Senioridade')
plt.ylabel('Salário em USD')
plt.show()
```

```python
import plotly.express as px
```

```python
df_avg_salary_plotly = df_limpo.groupby('experience_level')['salary_in_usd'].mean().reindex(ordem).reset_index()
fig = px.bar(df_avg_salary_plotly, x='experience_level', y='salary_in_usd',
             title='Salário Médio por Senioridade (Plotly)',
             labels={'experience_level': 'Senioridade', 'salary_in_usd': 'Salário Médio Anual (USD)'},
             category_orders={'experience_level': ordem})
fig.show()
```

```python
remoto_contagem = df_limpo['remote_ratio'].value_counts().reset_index()
remoto_contagem.columns = ['remote_ratio', 'quantidade']
fig = px.pie(remoto_contagem, names='remote_ratio', values='quantidade', title='Proporção dos tipos de trabalho', hole=0.5)
fig.update_traces( textinfo='percent+label')
fig.show()
```

```python
remoto_contagem = df_limpo['remote_ratio'].value_counts().reset_index()
remoto_contagem.columns = ['remote_ratio', 'quantidade']
fig = px.pie(remoto_contagem, names='remote_ratio', values='quantidade', title='Proporção dos tipos de trabalho')
fig.show()
```
